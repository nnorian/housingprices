```python
#visualizind relationships between coloumns and the price to see how each category influiences the price
import seaborn as sns
import matplotlib.pyplot as plt 

sns.scatterplot(x='area', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='bedrooms', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='bathrooms', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='stories', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='mainroad', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='guestroom', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='basement', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='hotwaterheating', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='airconditioning', y = target_variable, data= df)
plt.show()

sns.boxplot(x='parking', y=target_variable, data=df)
plt.xticks(rotation=90)
plt.show()

sns.scatterplot(x='prefarea', y = target_variable, data= df)
plt.show()

sns.scatterplot(x='furnishingstatus', y = target_variable, data= df)
plt.show()
```
